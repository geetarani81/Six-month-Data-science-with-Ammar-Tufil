Concept of Binning : cut ant colum in samall groups ALSO CALLED GROUPING 

```python
bins = [0,1,5,12,18,30,50,80]
labels = ['infants', 'toddlers' , 'kids ,' 'Teens' , 'youngs', 'middle', 'old']
```

use formula on data set to cut in slinces 

```python
df['binned_age'] = pd.cut(df['age'], bins = bins , labels = labels)
```
